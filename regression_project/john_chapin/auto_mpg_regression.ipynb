```python
%pylab inline
import pandas as pd
```

```python
names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin', 'name']

# Read data from CSV, using hard-coded column names. Drop rows with NaN values.
df = pd.read_csv("data/auto-mpg.data", delim_whitespace=True, na_values=['?'], names=names).dropna()

# Transform rows into dicts of explanatory variables
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']].T.to_dict().values()

# And hang on to the Y values
Y = df['mpg']
```

```python
from sklearn.feature_extraction import DictVectorizer

# Expand categorical explanatory variables
vectorizer = DictVectorizer()
X_v = vectorizer.fit_transform(X)
print X_v.
```

```python
from sklearn.cross_validation import train_test_split

# Split data into test/train groups
X_train, X_test, Y_train, Y_test = train_test_split(X_v, Y)

# Print the shape of the training data
print X_train.shape
print Y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

# Instantiate the LinearRegression model
regression = LinearRegression()
```

```python
# Fit and evaluate our model
regression.fit(X_train, Y_train)


print regression.score(X_test, Y_test)
```

```python
# Link real mpg to predicted mpg, then sort by real mpg
predictions = regression.predict(X_test)
linked = zip(Y_test, predictions)
linked.sort(key=lambda l: l[0])

# Line plot of real mpg, with scattered '+' of predicted mpg
r = range(0, len(Y_test))
plt.plot(r, numpy.array(linked)[:,[0]], 'b-', r, numpy.array(linked)[:,[1]], 'g+')
plt.xlabel("Instance (sorted by real mpg)")
plt.ylabel("Miles per Gallon")
plt.legend(("Real","Predicted"))
plt.show()
```
