### Let's get some results. 

Okay. Now that we have all this crazy data, what should we do with it? The fundamental question of data science!

What data do we have? 

Variables
- year

- years of experience
- wage

Categories
- ethnicity
- industry

- residence

Booleans
- union member?
- married? 
- health problem?

So a certain question we can ask is: how does variable A respond to variable B for a certain combination of categories and booleans? 

Let's write a simple function for quickly calculating wages vs year. 


```python
# need to add in ability to change x-variable

def generateGraph(category, specifier, boolCat, varX, varY):
    
    # create dataframe to use
    cols = ['yes','no']
    mdf = pd.DataFrame(columns=cols)
    
    # sort data by category specifier first
    catData = data[data[category] == specifier]
    
    # loop through and find mean values
    idx = 0
    for n in years:
        
        xData = catData[catData['year'] == n]
        
        yMean = xData[xData[boolCat] == 'yes'][varY].mean()
        nMean = xData[xData[boolCat] == 'no'][varY].mean()
        
        mdf.loc[idx,'yes'] = yMean
        mdf.loc[idx,'no'] = nMean
        
        idx += 1
    
    # plot graph
    x = np.arange(8)
    plt.plot(x,mdf['yes'])
    plt.plot(x,mdf['no'])
    plt.ylim(0,3)
    plt.xlabel('1980+')
    plt.ylabel('wage')


        
    
```

```python
generateGraph('industry','Mining','married','year','wage')
```

```python
generateGraph('occupation','Sales_Workers','union','year','wage')
```

```python
generateGraph('ethn','hisp','union','year','wage')
```

### Shouldn't schooling affect wages?

```python
cols = ['school','wage']
sWages = pd.DataFrame(columns=cols)

idx = 0
for y in school:
    
    wMean = data[(data['school'] == y) & (data['year'] == 1980)]['wage'].mean()
        
    sWages.loc[idx,'school'] = y
    sWages.loc[idx,'wage'] = wMean
    idx += 1


    
```
