```python
import pandas as pd
import numpy as np
```

```python
daydata = pd.read_csv('day.csv',parse_dates=True)
#note that the first column represents "instance", so we should remove that
del daydata['instant']
daydata
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(daydata['weekday'],daydata['cnt'], alpha=.2)
```

```python
print daydata.columns
plt.scatter(daydata['temp'],daydata['cnt'], alpha=.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
tempregression = LinearRegression()
# print daydata['cnt'].shape
# print daydata['temp'].shape
# print
# print daydata['cnt'].describe
# print daydata['temp'].describe

X = daydata['temp']
y = daydata['cnt']

# print np.shape(X)
X = np.reshape(X, (len(X), 1))
# print np.shape(X)
y = np.reshape(y, (len(y), 1))

X_train, X_test, y_train, y_test  = train_test_split(X, y)

tempregression.fit(X_train, y_train)
print tempregression.score(X_test, y_test)
print 'looks like this isn\'t a great predictor...'
```

```python
minx = min(X)
maxx = max(X)

miny = tempregression.predict(minx)
maxy = tempregression.predict(maxx)
```

```python
xx = np.linspace(minx, maxx, num=50)
xx = xx.reshape(len(xx),1)

plt.scatter(X,y, alpha=.2)
plt.scatter(xx, tempregression.predict(xx), c='r')   #predicted model
```
