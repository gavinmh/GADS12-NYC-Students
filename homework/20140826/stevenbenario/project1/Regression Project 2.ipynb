```python
print 'Do we think the ridership changes based on regular riders or ad-hoc?'

daysofweek = np.arange(0,7)

casualdaycounts= []
registereddaycounts = []
for i in daysofweek:
    casualdaycounts.append(daydata.casual[daydata['weekday'] == i].sum())
    registereddaycounts.append(daydata.registered[daydata['weekday'] == i].sum())
    
p_casual = plt.bar(daysofweek, casualdaycounts,width=.8, color='r', label='Casual Rides')
p_registered = plt.bar(daysofweek, registereddaycounts, width=.4, color='b', label='Registered Riders')
plt.title('Casual Rides and Registered Rides by Day of the Week')
plt.show()
```
