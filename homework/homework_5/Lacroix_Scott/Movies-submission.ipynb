#Partitioning and Transforming the data

```python
# create feature representations with CountVectorizer to convert corpus to a matrix of token counts. 
# import the CountVectorizer transformer from sklearn.feature_extraction.text
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
# Create an instance of CountVectorizer. Set the keyword argument `binary` to True.
binary_vectorizer = CountVectorizer(binary=True)
```

```python

# import the `train_test_split` convenience function from the `cross_validation` module.
from sklearn.cross_validation import train_test_split
#  split the data set into training and testing sets
# X_train, X_test, y_train, y_test = 
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Check X and y
#print X[:5]
#print y[:5]
# fit the `binary_vectorizer` on the training set
binary_vectorizer.fit(X)

#transform the training set
X_train = binary_vectorizer.transform(X_train)

# how many words are in the vocabulary?
#print len(binary_vectorizer.vocabulary_)
#show dimensions of the training set
#print X_train[:5].todense()

#transform the test set
X_test = binary_vectorizer.transform(X_test)
#show dimensions of the test set
print X_test[:5].todense()

# what are the dimensions of the design matrices?
print X_train.shape
print X_test.shape
```

```python
#training the model
# import the `LogisticRegression` class from the `linear_model` module.
from sklearn.linear_model import LogisticRegression

# create an instance of `LogisticRegression`
classifier = LogisticRegression()
# TODO fit the classifier on the training data
classifier.fit(X_train, y_train)
```

#Evaluate the model

```python
# evaluate the classifier. What is the default performance measure for `LogisticRegression`? 
classifier.score(X_test, y_test)
```

```python
# Let's evaluate the classifier using other performance measures.
# import the `recall_score`, `precision_score`, and `f1_score` functions from the `metrics` module.
from sklearn.metrics import recall_score, precision_score, f1_score

# make predictions for the test set
predictions = classifier.predict(X_test)
probabilities = classifier.predict_proba(X_test)
#for i, p in enumerate(predictions[:5]):
    #print p

for i, p in probabilities[:5]:
    print p
```

```python
# evaluate the classifier's precision, recall, and F1 score.
print 'recall' , recall_score(y_test, predictions)
print 'precision', precision_score(y_test, predictions)
print 'f1' , f1_score(y_test, predictions)
```
