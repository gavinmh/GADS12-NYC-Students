write several functions like the following to apply to the data frame that represent the data set using strings instead of floats. Use this to summarize your data:

def short_or_long(x): if x > 4: return 'long' else: return 'short'

```python
def sepal_long_short(x): 
    if x > 5.2: return 'long' 
    else: return 'short'
    
def sepal_large_narrow(x): 
    if x > 3.0: return 'large' 
    else: return 'narrow'
    
def petal_long_short(x): 
    if x >= 4.5: return 'long' 
    else: return 'short'
    
def petal_large_narrow(x): 
    if x >= 1.4: return 'large' 
    else: return 'narrow'
```

CHALLENGE QUESTION: From everything above, you should be able to predict relatively accurately for each row what kind of flower it is (without using the Name column as an obvious hint). Write a function that uses the data with if and else statements to attempt to classify each row.

```python
def guess_flower(row, debug=0):
    pLength = petal_long_short(row['PetalLength'])
    pWidth = petal_large_narrow(row['PetalWidth'])
    sLength = sepal_long_short(row['SepalLength'])
    sWidth = sepal_large_narrow(row['SepalWidth'])
    guess = 'Iris-versicolor' # by default the 'medium' flower
    
    if pLength == 'long' and pWidth == 'large':
            guess = 'Iris-virginica' # if petal is long and large
    elif sWidth == 'large':
        guess = 'Iris-setosa' # if sepal is large

    if debug:
        print str(row['Name'] == guess).upper(), ': ', row['Name'],  ' -vs- ', guess, \
        ' -> ', str(row['PetalLength']), '/', str(row['PetalWidth']), \
        '/', str(row['SepalLength']), '/', str(row['SepalWidth'])

    # return True if actual name is same as the guessed name
    return row['Name'] == guess
```

```python
guesses = []
for i,row in data.iterrows():
    guesses += [guess_flower(row)]
print '%.1f%% of correct guesses' % (100.0 * sum(guesses) / len(guesses))
```
