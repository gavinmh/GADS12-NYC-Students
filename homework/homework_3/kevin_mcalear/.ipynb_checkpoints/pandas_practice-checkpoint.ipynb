```python
# Importing Pandas
import pandas as pd
```

```python
# Pull in the data from the website
data = pd.read_csv('https://raw.github.com/pydata/pandas/master/pandas/tests/data/iris.csv')

#show the first five entries to make sure it worked
data.head()
```

```python
# How many different names of flowers are in this data set?
print data['Name'].unique()
print data['Name'].nunique()
```

```python
# How many exist of each type of flower?
print str(len(data[data['Name'] == 'Iris-setosa'])) + ' Iris-setosas'
print str(len(data[data['Name'] == 'Iris-versicolor'])) + ' Iris-versicolor'
print str(len(data[data['Name'] == 'Iris-virginica'])) + ' Iris-virginica'
```

```python
# Determine the min, median, mean, max for each numeric feature in the data set.
data.describe()
```

```python
# Determine the same for each flower type.
print 'Iris-setosa Stats:'
print '-------------------------------------------------------'
print data[data['Name'] == 'Iris-setosa'].describe()
print '-------------------------------------------------------'
print
print
print 'Iris-versicolor Stats:'
print '-------------------------------------------------------'
print data[data['Name'] == 'Iris-versicolor'].describe()
print '-------------------------------------------------------'
print
print
print 'Iris-virginica Stats:'
print '-------------------------------------------------------'
print data[data['Name'] == 'Iris-virginica'].describe()
print '-------------------------------------------------------'
```
