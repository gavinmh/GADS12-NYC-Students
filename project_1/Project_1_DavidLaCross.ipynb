#Art Sales Data
This project examines auction records from 9 contemporary artists. The artists in the data set include:

| Artist               | Record Count |
|----------------------|:------------:|
| Andy Warhol          |      445     |
| Ed Ruscha            |      92      |
| Ellsworth Kelly      |      72      |
| Frank Stella         |      89      |
| George Condo         |      62      |
| Jean-Michel Basquiat |      58      |
| Robert Rauschenberg  |      90      |
| Roy Lichtenstein     |      143     |
| Tom Wesselmann       |      96      |

Each record contains the following attributes:

| Field         |   Type  | Description                                            |
|---------------|:-------:|--------------------------------------------------------|
| auction_house |  String | The name of the auction house conducting the sale      |
| artist_name   |  String | Name of the artist                                     |
| low_est       | integer | The lower limit of the estimated sale price            |
| hi_est        | integer | The upper limit of the estimated sale price            |
| sale_price    | integer | The sale price realized at auction                     |
| currency      |  string | For this dataset, always USD                           |
| quarter       |  string | The quarter of the year during which the sale occurred |
| time          |  string | Afternoon vs. evening sales                            |

```python
# Start by loading the required libraries and getting the art sales data into a dataframe

import pandas as pd
import numpy as np

auction_data = pd.read_csv('data/auction_data_refined.csv')
```

```python
# looking at the variance in the numeric values:

var_saleprice =  auction_data['sale_price'].var()
var_loest =  auction_data['lo_est'].var()
var_hiest =  auction_data['hi_est'].var()
print "The variance values are:"
print "Sale price: " + str(var_saleprice)
print "Low Estimate: " + str(var_loest)
print "High Estimate: " + str(var_hiest)
```

```python
# split the current dataframe into two: one containing the explanatory variables and one with the response

data = auction_data[["artist_name", "auction_house", "quarter", "time", "lo_est", "hi_est"]]
response = auction_data[["sale_price"]]

print data.head()
print response.head()

print data.shape
print response.shape
```

```python
# need to make the explanatory dataframe into a dict so the strings
# can be vectorized by DictVectorizer

data_dict = data.to_dict('records')
print data_dict
```

```python
# import and instantiate a vecorizer

from sklearn.feature_extraction import DictVectorizer

v = DictVectorizer(sparse=False)
data_vector = v.fit_transform(data_dict)

print type(data_vector)
print data_vector.shape
```

```python
# split everything into training and test sets

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_vector, response, test_size=0.33)

# print the shape of the new arrays to confirm it looks right
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
# start training the model

# import LinearRegression from the linear_model module of sklearn
from sklearn.linear_model import LinearRegression

# instantiate a regressor and fit the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```
